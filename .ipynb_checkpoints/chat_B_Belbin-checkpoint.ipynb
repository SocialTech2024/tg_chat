```python
# словарь с ролями и ключевыми словами
roles_keywords = {"Координатор": [
        "организовать", "планировать", "координировать", "назначить встречу", "распределить задачи",
        "назначить ответственных", "управлять командой", "обсудить план", "разработать стратегию",
        "назначить крайний срок", "встреча", "руководить", "контролировать процесс", "курировать проект",
        "ответственность за команду", "координировать действия", "плановая встреча", "стратегическое планирование",
        "организация мероприятий", "составить расписание", "делегировать задачи", "контроль исполнения",
        "установить цели", "провести собрание", "обсудить результаты", "координация команды", "планировать сроки",
        "мониторинг выполнения", "согласование действий", "информировать команду", "управлять проектом",
        "организовать ресурсы", "выделить приоритеты", "обеспечить взаимодействие", "управлять изменениями",
        "оценить потребности", "формировать команды", "обеспечить прозрачность", "налаживать коммуникацию",
        "поддерживать мотивацию команды", "управлять конфликтами", "анализировать процессы", "выстраивать взаимоотношения",
        "разрабатывать планы действий", "координировать проекты", "подготавливать отчеты", "определять ключевые показатели",
        "внедрять процессы"
    ],
    "Исполнитель": [
        "выполнить", "начать", "завершить", "реализовать", "привести в действие", "довести до конца",
        "закончить", "приступить к работе", "следовать плану", "выполнить задачу", "разработать",
        "выполнение проекта", "завершение работы", "взяться за дело", "решить задачу", "достичь результата",
        "осуществить", "внедрить", "исполнить", "воплотить в жизнь", "практическое выполнение",
        "реализация плана", "сделать", "оперативное выполнение", "план", "сроки", "достигнуть",
        "завершение этапа", "выполнение требований", "обеспечить качество", "работать по графику",
        "поддерживать стандарт", "исполнять обязанности", "отвечать за результаты", "реализовывать задачи",
        "действовать согласно плану", "выдвигать предложения", "докладывать о прогрессе"
    ],
    "Мотиватор": [
        "вдохновить", "мотивировать", "поддержать", "воодушевить", "подбодрить", "двигаться вперед",
        "стимулировать", "ускориться", "показать результат", "стремиться", "давайте начнем",
        "мы можем это сделать", "добиться успеха", "сделать быстро", "продолжать движение",
        "направить усилия", "подтолкнуть команду", "вдохновение", "активировать действия",
        "повысить результат", "энергия для команды", "продвижение", "развить усилия",
        "мотивировать на успех", "призыв к действию", "воодушевление", "стимул", "энтузиазм",
        "создать позитивный климат", "обсуждение целей", "достигать вместе", "поддерживать командный дух",
        "устанавливать амбициозные цели", "поощрять инициативу", "развивать уверенность", "вдохновлять на новые идеи",
        "создавать мотивационную атмосферу", "обсуждать достижения", "поддерживать активность команды"
    ],
    "Оценщик": [
        "оценить", "проанализировать", "взвесить риски", "рассмотреть варианты", "анализировать",
        "взвешивание плюсов и минусов", "критический взгляд", "оценка ситуации", "предусмотреть последствия",
        "принять во внимание риски", "сравнить варианты", "анализировать результат", "аналитический подход",
        "оценить эффективность", "анализировать проект", "критический анализ", "предсказать исход",
        "оценить выгоду", "критическая оценка", "взвесить факты", "анализ данных", "тщательный анализ",
        "оценка рисков", "проверить варианты", "оценить последствия", "сделать выводы", "исследовать",
        "детали", "проверить", "качество", "обратить внимание", "анализ трендов", "выявить слабые места",
        "оценка результатов", "анализировать альтернативы", "оценить целесообразность", "выявить возможности",
        "проводить SWOT-анализ", "проверка соответствия", "сравнение показателей", "оценка стоимости",
        "анализ причинно-следственных связей"
    ],
    "Исследователь ресурсов": [
        "найти ресурсы", "поискать возможности", "исследовать", "разведать", "выявить новые возможности",
        "найти контакты", "налаживать связи", "найти решения", "привлечь внешние ресурсы",
        "обнаружить новые пути", "возможности сотрудничества", "узнать больше", "поиск идей",
        "найти новые идеи", "изучить варианты", "установить контакты", "партнерство",
        "поиск новых партнеров", "поиск информации", "новые ресурсы", "изучение рынка",
        "расширить контакты", "привлечение инвестиций", "исследование рынка", "поиск поддержки",
        "расширение возможностей", "анализ конкурентов", "разработка партнерств",
        "поиск инновационных решений", "создание сетей", "мониторинг тенденций",
        "выявление потенциальных ресурсов", "оценка возможностей", "оптимизация процессов",
        "сбор информации", "координация с заинтересованными сторонами",
        "выстраивание стратегий сотрудничества", "исследование новых направлений"
    ],
    "Творец": [
        "придумать", "создать", "предложить идею", "разработать новый подход", "генерировать идеи",
        "творческое решение", "нестандартное решение", "изобрести", "придумать новое",
        "применить креатив", "креативное мышление", "создать концепт", "воплотить идею",
        "разработать инновацию", "новаторский подход", "предложить нестандартный путь", "новая концепция",
        "креативное решение", "генерация идей", "найти оригинальное решение", "разработать прототип",
        "предложить новую стратегию", "инновационное решение", "создание концепции", "креатив",
        "новаторство", "идея", "дизайн решения", "экспериментировать", "создать новое направление",
        "внедрить инновации", "разрабатывать новые подходы", "исследовать новые возможности",
        "создать уникальное решение", "креативный подход", "интерпретировать идеи",
        "развивать креативные концепции"
    ],
    "Командный игрок": [
        "поддержать команду", "помочь", "работать вместе", "совместная работа", "взаимодействовать",
        "помочь коллеге", "сотрудничать", "внести вклад в команду", "объединить усилия",
        "помочь решить задачу", "вместе решим", "поддерживать других", "укрепление команды",
        "сглаживание конфликтов", "поддержать коллег", "сплочение команды", "вместе к успеху",
        "решение конфликтов", "создание единства", "командная работа", "взаимопомощь",
        "объединение усилий", "сплоченность коллектива", "вклад в общую задачу", "поддержка в команде",
        "создание команды", "совместная деятельность", "взаимопонимание", "командный дух",
        "развитие отношений", "поддержка инициатив", "культура сотрудничества",
        "активное участие в группе", "обсуждение идей", "разрешение разногласий",
        "взаимная поддержка", "коллективное принятие решений", "сплоченность группы",
        "разделение ответственности", "построение доверительных отношений"
    ],
    "Завершитель": [
        "закончить вовремя", "контролировать сроки", "точность", "дедлайн", "соблюдать сроки",
        "успеть вовремя", "достичь результата в срок", "следить за сроками", "финальный контроль",
        "вовремя завершить", "отслеживание прогресса", "выполнение задачи в срок", "контроль качества",
        "точная реализация", "закрыть проект", "завершение проекта", "выполнение обязательств",
        "достижение цели вовремя", "своевременное завершение", "сроки сдачи", "контроль выполнения",
        "отслеживание выполнения", "мониторинг дедлайнов", "тщательный контроль", "следить за процессом",
        "вовремя выполнить", "точное завершение", "дедлайн проекта", "финализация", "управление сроками",
        "отчетность по срокам", "фокус на результатах", "анализ сроков", "обеспечить соблюдение сроков",
        "планирование временных рамок", "документирование прогресса", "регулярные отчеты о статусе",
        "актуализация планов", "корректировка сроков", "достижение ключевых этапов"
    ],
    "Специалист": [
        "эксперт", "знания", "компетентность", "специализация", "профессиональные навыки",
        "глубокие знания", "опыт", "эксперт в области", "высокая квалификация", "специалист",
        "специальные знания", "опыт работы", "глубокое понимание", "уникальные навыки",
        "профессионализм", "экспертные советы", "технические знания", "глубокая экспертиза",
        "узкая специализация", "навыки эксперта", "профессиональный опыт", "технические навыки",
        "навыки в области", "глубокая специализация", "специализированные знания",
        "специалист в своей области", "экспертное заключение", "экспертные знания",
        "профессиональные советы", "компетенция", "экспертиза", "глубокое понимание процессов",
        "экспертный анализ", "профессиональная консультация", "анализ данных", "практические навыки",
        "методологические знания", "навыки решения задач", "применение лучших практик",
        "глубокое понимание технологий", "постоянное обучение", "инновационные подходы",
        "профессиональные стандарты"
    ]
}


# Функция для поиска ролей в сообщениях
def identify_roles(message):
    identified_roles = []

    # Преобразуем message в строку, если это не строка
    if not isinstance(message, str):
        message = str(message)

    for role, keywords in roles_keywords.items():
        for keyword in keywords:
            if keyword in message:
                identified_roles.append(role)
                break  # Выходим из цикла, если нашли соответствие
    return identified_roles

file_path = '/content/clear_Result_file.csv'  # Укажите путь к вашему CSV файлу
data = pd.read_csv(file_path)

# сообщения находятся в столбце 'text'
if 'text' in data.columns:
    # Идентификация ролей в сообщениях
    for index, row in data.iterrows():
        message = row['text']
        roles = identify_roles(message)
        # Выводим только определенные роли
        if roles:  # Если роли найдены
            print(f"Сообщение: '{message}'\nРоли: {roles}\n")
else:
    print("Столбец 'text' не найден в файле.")
```

```python
import pandas as pd
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize

# Загрузка необходимых ресурсов
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Загрузка данных из CSV файла
df = pd.read_csv('/content/clear_Result_file.csv')

# Создание DataFrame
# Необязательно, если Вы уже загружаете DataFrame из CSV
# df = pd.DataFrame(data)

# Функция для извлечения глаголов и существительных
def extract_nouns_verbs(text):
    if isinstance(text, str):  # Проверка, является ли текст строкой
        words = word_tokenize(text)  # Токенизация
        pos_tags = pos_tag(words)     # Определение частей речи
        nouns_verbs = [word for word, pos in pos_tags if pos.startswith('N') or pos.startswith('V')]
        return nouns_verbs
    return []  # Возврат пустого списка, если текст не строка

# Применение функции ко всем текстам в DataFrame
all_nouns_verbs = []
for text in df['text']:
    nouns_verbs = extract_nouns_verbs(text)
    all_nouns_verbs.extend(nouns_verbs)

# Вывод результатов
print(set(all_nouns_verbs))  # Уникальные слова
```

```python
import nltk
import csv
from nltk.tokenize import word_tokenize

# Исходный словарь с ролями и ключевыми словами
roles_keywords = {"Координатор": [
        "организовать", "планировать", "координировать", "назначить встречу", "распределить задачи",
        "назначить ответственных", "управлять командой", "обсудить план", "разработать стратегию",
        "назначить крайний срок", "встреча", "руководить", "контролировать процесс", "курировать проект",
        "ответственность за команду", "координировать действия", "плановая встреча", "стратегическое планирование",
        "организация мероприятий", "составить расписание", "делегировать задачи", "контроль исполнения",
        "установить цели", "провести собрание", "обсудить результаты", "координация команды", "планировать сроки",
        "мониторинг выполнения", "согласование действий", "информировать команду", "управлять проектом",
        "организовать ресурсы", "выделить приоритеты", "обеспечить взаимодействие", "управлять изменениями",
        "оценить потребности", "формировать команды", "обеспечить прозрачность", "налаживать коммуникацию",
        "поддерживать мотивацию команды", "управлять конфликтами", "анализировать процессы", "выстраивать взаимоотношения",
        "разрабатывать планы действий", "координировать проекты", "подготавливать отчеты", "определять ключевые показатели",
        "внедрять процессы"
    ],
    "Исполнитель": [
        "выполнить", "начать", "завершить", "реализовать", "привести в действие", "довести до конца",
        "закончить", "приступить к работе", "следовать плану", "выполнить задачу", "разработать",
        "выполнение проекта", "завершение работы", "взяться за дело", "решить задачу", "достичь результата",
        "осуществить", "внедрить", "исполнить", "воплотить в жизнь", "практическое выполнение",
        "реализация плана", "сделать", "оперативное выполнение", "план", "сроки", "достигнуть",
        "завершение этапа", "выполнение требований", "обеспечить качество", "работать по графику",
        "поддерживать стандарт", "исполнять обязанности", "отвечать за результаты", "реализовывать задачи",
        "действовать согласно плану", "выдвигать предложения", "докладывать о прогрессе"
    ],
    "Мотиватор": [
        "вдохновить", "мотивировать", "поддержать", "воодушевить", "подбодрить", "двигаться вперед",
        "стимулировать", "ускориться", "показать результат", "стремиться", "давайте начнем",
        "мы можем это сделать", "добиться успеха", "сделать быстро", "продолжать движение",
        "направить усилия", "подтолкнуть команду", "вдохновение", "активировать действия",
        "повысить результат", "энергия для команды", "продвижение", "развить усилия",
        "мотивировать на успех", "призыв к действию", "воодушевление", "стимул", "энтузиазм",
        "создать позитивный климат", "обсуждение целей", "достигать вместе", "поддерживать командный дух",
        "устанавливать амбициозные цели", "поощрять инициативу", "развивать уверенность", "вдохновлять на новые идеи",
        "создавать мотивационную атмосферу", "обсуждать достижения", "поддерживать активность команды"
    ],
    "Оценщик": [
        "оценить", "проанализировать", "взвесить риски", "рассмотреть варианты", "анализировать",
        "взвешивание плюсов и минусов", "критический взгляд", "оценка ситуации", "предусмотреть последствия",
        "принять во внимание риски", "сравнить варианты", "анализировать результат", "аналитический подход",
        "оценить эффективность", "анализировать проект", "критический анализ", "предсказать исход",
        "оценить выгоду", "критическая оценка", "взвесить факты", "анализ данных", "тщательный анализ",
        "оценка рисков", "проверить варианты", "оценить последствия", "сделать выводы", "исследовать",
        "детали", "проверить", "качество", "обратить внимание", "анализ трендов", "выявить слабые места",
        "оценка результатов", "анализировать альтернативы", "оценить целесообразность", "выявить возможности",
        "проводить SWOT-анализ", "проверка соответствия", "сравнение показателей", "оценка стоимости",
        "анализ причинно-следственных связей"
    ],
    "Исследователь ресурсов": [
        "найти ресурсы", "поискать возможности", "исследовать", "разведать", "выявить новые возможности",
        "найти контакты", "налаживать связи", "найти решения", "привлечь внешние ресурсы",
        "обнаружить новые пути", "возможности сотрудничества", "узнать больше", "поиск идей",
        "найти новые идеи", "изучить варианты", "установить контакты", "партнерство",
        "поиск новых партнеров", "поиск информации", "новые ресурсы", "изучение рынка",
        "расширить контакты", "привлечение инвестиций", "исследование рынка", "поиск поддержки",
        "расширение возможностей", "анализ конкурентов", "разработка партнерств",
        "поиск инновационных решений", "создание сетей", "мониторинг тенденций",
        "выявление потенциальных ресурсов", "оценка возможностей", "оптимизация процессов",
        "сбор информации", "координация с заинтересованными сторонами",
        "выстраивание стратегий сотрудничества", "исследование новых направлений"
    ],
    "Творец": [
        "придумать", "создать", "предложить идею", "разработать новый подход", "генерировать идеи",
        "творческое решение", "нестандартное решение", "изобрести", "придумать новое",
        "применить креатив", "креативное мышление", "создать концепт", "воплотить идею",
        "разработать инновацию", "новаторский подход", "предложить нестандартный путь", "новая концепция",
        "креативное решение", "генерация идей", "найти оригинальное решение", "разработать прототип",
        "предложить новую стратегию", "инновационное решение", "создание концепции", "креатив",
        "новаторство", "идея", "дизайн решения", "экспериментировать", "создать новое направление",
        "внедрить инновации", "разрабатывать новые подходы", "исследовать новые возможности",
        "создать уникальное решение", "креативный подход", "интерпретировать идеи",
        "развивать креативные концепции"
    ],
    "Командный игрок": [
        "поддержать команду", "помочь", "работать вместе", "совместная работа", "взаимодействовать",
        "помочь коллеге", "сотрудничать", "внести вклад в команду", "объединить усилия",
        "помочь решить задачу", "вместе решим", "поддерживать других", "укрепление команды",
        "сглаживание конфликтов", "поддержать коллег", "сплочение команды", "вместе к успеху",
        "решение конфликтов", "создание единства", "командная работа", "взаимопомощь",
        "объединение усилий", "сплоченность коллектива", "вклад в общую задачу", "поддержка в команде",
        "создание команды", "совместная деятельность", "взаимопонимание", "командный дух",
        "развитие отношений", "поддержка инициатив", "культура сотрудничества",
        "активное участие в группе", "обсуждение идей", "разрешение разногласий",
        "взаимная поддержка", "коллективное принятие решений", "сплоченность группы",
        "разделение ответственности", "построение доверительных отношений"
    ],
    "Завершитель": [
        "закончить вовремя", "контролировать сроки", "точность", "дедлайн", "соблюдать сроки",
        "успеть вовремя", "достичь результата в срок", "следить за сроками", "финальный контроль",
        "вовремя завершить", "отслеживание прогресса", "выполнение задачи в срок", "контроль качества",
        "точная реализация", "закрыть проект", "завершение проекта", "выполнение обязательств",
        "достижение цели вовремя", "своевременное завершение", "сроки сдачи", "контроль выполнения",
        "отслеживание выполнения", "мониторинг дедлайнов", "тщательный контроль", "следить за процессом",
        "вовремя выполнить", "точное завершение", "дедлайн проекта", "финализация", "управление сроками",
        "отчетность по срокам", "фокус на результатах", "анализ сроков", "обеспечить соблюдение сроков",
        "планирование временных рамок", "документирование прогресса", "регулярные отчеты о статусе",
        "актуализация планов", "корректировка сроков", "достижение ключевых этапов"
    ],
    "Специалист": [
        "эксперт", "знания", "компетентность", "специализация", "профессиональные навыки",
        "глубокие знания", "опыт", "эксперт в области", "высокая квалификация", "специалист",
        "специальные знания", "опыт работы", "глубокое понимание", "уникальные навыки",
        "профессионализм", "экспертные советы", "технические знания", "глубокая экспертиза",
        "узкая специализация", "навыки эксперта", "профессиональный опыт", "технические навыки",
        "навыки в области", "глубокая специализация", "специализированные знания",
        "специалист в своей области", "экспертное заключение", "экспертные знания",
        "профессиональные советы", "компетенция", "экспертиза", "глубокое понимание процессов",
        "экспертный анализ", "профессиональная консультация", "анализ данных", "практические навыки",
        "методологические знания", "навыки решения задач", "применение лучших практик",
        "глубокое понимание технологий", "постоянное обучение", "инновационные подходы",
        "профессиональные стандарты"
    ]
}


# Функция для токенизации ключевых слов по каждой роли
def tokenize_keywords(roles_keywords):
    tokenized_roles_keywords = {}
    for role, keywords in roles_keywords.items():
        tokenized_keywords = list(set(word_tokenize(" ".join(keywords))))
        tokenized_roles_keywords[role] = tokenized_keywords
    return tokenized_roles_keywords

# Загрузка nltk словарей (выполните один раз)
nltk.download('punkt')

# Токенизация словаря с ролями и ключевыми словами
tokenized_roles_keywords = tokenize_keywords(roles_keywords)

# Сохранение токенизированного словаря в CSV-файл
with open('slovar_tokenized_roles_keywords.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Role", "Keywords"])  # Заголовки столбцов
    for role, keywords in tokenized_roles_keywords.items():
        writer.writerow([role, ", ".join(keywords)])

print("Токенизированный словарь успешно сохранен в 'tokenized_roles_keywords.csv'")
```

```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertModel

# 1. Загрузка JSON в DataFrame
df = pd.read_json('/content/result_chat3.json')

# 2. Проверка структуры данных
print(df.head())  # Проверка первых нескольких строк DataFrame
print(df.columns)  # Вывод всех колонок DataFrame

# 3. Убедитесь, что колонка 'text' существует и является строкой
if 'messages' not in df.columns:
    raise ValueError("В Вашем JSON файле нет колонки 'text'.")

df['messages'] = df['messages'].astype(str)

# 4. Загрузка модели и токенизатора
model_name = 'DeepPavlov/rubert-base-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# Функция для получения эмбеддингов
def get_embeddings(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    with torch.no_grad():  # отключение градиентов
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

# Применение функции к колонке 'text', чтобы получить эмбеддинги
df['embeddings'] = df['messages'].apply(get_embeddings)

# Проверка результата
print(df[['messages', 'embeddings']])
```

```python
import json
import pandas as pd
import torch
from transformers import BertTokenizer, BertModel

# 1. Загрузка JSON-данных
with open('/content/result_chat3.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# 2. Извлечение сообщений
messages = data['messages']

# 3. Подготовка списка текстов
texts = []

for message in messages:
    if message['type'] == 'message':
        # Смотрим, если текст - это список
        if isinstance(message['text'], list):
            # Объединяем все части текста в одну строку
            full_text = ''.join([t['text'] if isinstance(t, dict) else t for t in message['text']])
            texts.append(full_text)
        else:
            texts.append(message['text'])  # Если текст не в виде списка

# 4. Преобразование в DataFrame
df = pd.DataFrame(texts, columns=['text'])

# 5. Загрузка модели и токенизатора
model_name = 'DeepPavlov/rubert-base-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# Убедитесь, что Ваша модель на GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# 6. Функция для получения эмбеддингов
def get_embeddings(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()

# 7. Применение функции к колонке 'text' для получения эмбеддингов
df['embeddings'] = df['text'].apply(get_embeddings)

# 8. Проверка результата
print(df[['text', 'embeddings']].head())

# 9. Сохранение DataFrame с эмбеддингами в CSV (если необходимо)
df.to_csv('/content/embeddings_result.csv', index=False)
```

```python
# Загрузка модели и токенизатора
model_name = 'DeepPavlov/rubert-base-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)


roles_keywords = {"Координатор": [
        "организовать", "планировать", "координировать", "назначить встречу", "распределить задачи",
        "назначить ответственных", "управлять командой", "обсудить план", "разработать стратегию",
        "назначить крайний срок", "встреча", "руководить", "контролировать процесс", "курировать проект",
        "ответственность за команду", "координировать действия", "плановая встреча", "стратегическое планирование",
        "организация мероприятий", "составить расписание", "делегировать задачи", "контроль исполнения",
        "установить цели", "провести собрание", "обсудить результаты", "координация команды", "планировать сроки",
        "мониторинг выполнения", "согласование действий", "информировать команду", "управлять проектом",
        "организовать ресурсы", "выделить приоритеты", "обеспечить взаимодействие", "управлять изменениями",
        "оценить потребности", "формировать команды", "обеспечить прозрачность", "налаживать коммуникацию",
        "поддерживать мотивацию команды", "управлять конфликтами", "анализировать процессы", "выстраивать взаимоотношения",
        "разрабатывать планы действий", "координировать проекты", "подготавливать отчеты", "определять ключевые показатели",
        "внедрять процессы"
    ],
    "Исполнитель": [
        "выполнить", "начать", "завершить", "реализовать", "привести в действие", "довести до конца",
        "закончить", "приступить к работе", "следовать плану", "выполнить задачу", "разработать",
        "выполнение проекта", "завершение работы", "взяться за дело", "решить задачу", "достичь результата",
        "осуществить", "внедрить", "исполнить", "воплотить в жизнь", "практическое выполнение",
        "реализация плана", "сделать", "оперативное выполнение", "план", "сроки", "достигнуть",
        "завершение этапа", "выполнение требований", "обеспечить качество", "работать по графику",
        "поддерживать стандарт", "исполнять обязанности", "отвечать за результаты", "реализовывать задачи",
        "действовать согласно плану", "выдвигать предложения", "докладывать о прогрессе"
    ],
    "Мотиватор": [
        "вдохновить", "мотивировать", "поддержать", "воодушевить", "подбодрить", "двигаться вперед",
        "стимулировать", "ускориться", "показать результат", "стремиться", "давайте начнем",
        "мы можем это сделать", "добиться успеха", "сделать быстро", "продолжать движение",
        "направить усилия", "подтолкнуть команду", "вдохновение", "активировать действия",
        "повысить результат", "энергия для команды", "продвижение", "развить усилия",
        "мотивировать на успех", "призыв к действию", "воодушевление", "стимул", "энтузиазм",
        "создать позитивный климат", "обсуждение целей", "достигать вместе", "поддерживать командный дух",
        "устанавливать амбициозные цели", "поощрять инициативу", "развивать уверенность", "вдохновлять на новые идеи",
        "создавать мотивационную атмосферу", "обсуждать достижения", "поддерживать активность команды"
    ],
    "Оценщик": [
        "оценить", "проанализировать", "взвесить риски", "рассмотреть варианты", "анализировать",
        "взвешивание плюсов и минусов", "критический взгляд", "оценка ситуации", "предусмотреть последствия",
        "принять во внимание риски", "сравнить варианты", "анализировать результат", "аналитический подход",
        "оценить эффективность", "анализировать проект", "критический анализ", "предсказать исход",
        "оценить выгоду", "критическая оценка", "взвесить факты", "анализ данных", "тщательный анализ",
        "оценка рисков", "проверить варианты", "оценить последствия", "сделать выводы", "исследовать",
        "детали", "проверить", "качество", "обратить внимание", "анализ трендов", "выявить слабые места",
        "оценка результатов", "анализировать альтернативы", "оценить целесообразность", "выявить возможности",
        "проводить SWOT-анализ", "проверка соответствия", "сравнение показателей", "оценка стоимости",
        "анализ причинно-следственных связей"
    ],
    "Исследователь ресурсов": [
        "найти ресурсы", "поискать возможности", "исследовать", "разведать", "выявить новые возможности",
        "найти контакты", "налаживать связи", "найти решения", "привлечь внешние ресурсы",
        "обнаружить новые пути", "возможности сотрудничества", "узнать больше", "поиск идей",
        "найти новые идеи", "изучить варианты", "установить контакты", "партнерство",
        "поиск новых партнеров", "поиск информации", "новые ресурсы", "изучение рынка",
        "расширить контакты", "привлечение инвестиций", "исследование рынка", "поиск поддержки",
        "расширение возможностей", "анализ конкурентов", "разработка партнерств",
        "поиск инновационных решений", "создание сетей", "мониторинг тенденций",
        "выявление потенциальных ресурсов", "оценка возможностей", "оптимизация процессов",
        "сбор информации", "координация с заинтересованными сторонами",
        "выстраивание стратегий сотрудничества", "исследование новых направлений"
    ],
    "Творец": [
        "придумать", "создать", "предложить идею", "разработать новый подход", "генерировать идеи",
        "творческое решение", "нестандартное решение", "изобрести", "придумать новое",
        "применить креатив", "креативное мышление", "создать концепт", "воплотить идею",
        "разработать инновацию", "новаторский подход", "предложить нестандартный путь", "новая концепция",
        "креативное решение", "генерация идей", "найти оригинальное решение", "разработать прототип",
        "предложить новую стратегию", "инновационное решение", "создание концепции", "креатив",
        "новаторство", "идея", "дизайн решения", "экспериментировать", "создать новое направление",
        "внедрить инновации", "разрабатывать новые подходы", "исследовать новые возможности",
        "создать уникальное решение", "креативный подход", "интерпретировать идеи",
        "развивать креативные концепции"
    ],
    "Командный игрок": [
        "поддержать команду", "помочь", "работать вместе", "совместная работа", "взаимодействовать",
        "помочь коллеге", "сотрудничать", "внести вклад в команду", "объединить усилия",
        "помочь решить задачу", "вместе решим", "поддерживать других", "укрепление команды",
        "сглаживание конфликтов", "поддержать коллег", "сплочение команды", "вместе к успеху",
        "решение конфликтов", "создание единства", "командная работа", "взаимопомощь",
        "объединение усилий", "сплоченность коллектива", "вклад в общую задачу", "поддержка в команде",
        "создание команды", "совместная деятельность", "взаимопонимание", "командный дух",
        "развитие отношений", "поддержка инициатив", "культура сотрудничества",
        "активное участие в группе", "обсуждение идей", "разрешение разногласий",
        "взаимная поддержка", "коллективное принятие решений", "сплоченность группы",
        "разделение ответственности", "построение доверительных отношений"
    ],
    "Завершитель": [
        "закончить вовремя", "контролировать сроки", "точность", "дедлайн", "соблюдать сроки",
        "успеть вовремя", "достичь результата в срок", "следить за сроками", "финальный контроль",
        "вовремя завершить", "отслеживание прогресса", "выполнение задачи в срок", "контроль качества",
        "точная реализация", "закрыть проект", "завершение проекта", "выполнение обязательств",
        "достижение цели вовремя", "своевременное завершение", "сроки сдачи", "контроль выполнения",
        "отслеживание выполнения", "мониторинг дедлайнов", "тщательный контроль", "следить за процессом",
        "вовремя выполнить", "точное завершение", "дедлайн проекта", "финализация", "управление сроками",
        "отчетность по срокам", "фокус на результатах", "анализ сроков", "обеспечить соблюдение сроков",
        "планирование временных рамок", "документирование прогресса", "регулярные отчеты о статусе",
        "актуализация планов", "корректировка сроков", "достижение ключевых этапов"
    ],
    "Специалист": [
        "эксперт", "знания", "компетентность", "специализация", "профессиональные навыки",
        "глубокие знания", "опыт", "эксперт в области", "высокая квалификация", "специалист",
        "специальные знания", "опыт работы", "глубокое понимание", "уникальные навыки",
        "профессионализм", "экспертные советы", "технические знания", "глубокая экспертиза",
        "узкая специализация", "навыки эксперта", "профессиональный опыт", "технические навыки",
        "навыки в области", "глубокая специализация", "специализированные знания",
        "специалист в своей области", "экспертное заключение", "экспертные знания",
        "профессиональные советы", "компетенция", "экспертиза", "глубокое понимание процессов",
        "экспертный анализ", "профессиональная консультация", "анализ данных", "практические навыки",
        "методологические знания", "навыки решения задач", "применение лучших практик",
        "глубокое понимание технологий", "постоянное обучение", "инновационные подходы",
        "профессиональные стандарты"
    ]
}
# Преобразование в DataFrame
data = []
for role, keywords in roles_keywords.items():
    for keyword in keywords:
        data.append({"Role": role, "Keyword": keyword})

df_roles = pd.DataFrame(data)

# 2. Загрузка модели и токенизатора BERT
model_name = 'DeepPavlov/rubert-base-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# 3. Функция для получения эмбеддингов
def get_keyword_embeddings(keyword):
    inputs = tokenizer(keyword, return_tensors='pt', padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

# 4. Применение эмбеддингов к колонке 'Keyword'
df_roles['Embeddings'] = df_roles['Keyword'].apply(get_keyword_embeddings)

# 5. Сохранение в CSV файл
df_roles.to_csv('roles_keywords_embeddings.csv', index=False)

```
